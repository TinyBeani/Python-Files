```python
def gradient_descent(x0, y0, f, grad_f, alpha, num_iterations):
    
    x, y = x0, y0  # Initial point
    for i in range(num_iterations):#gradient of f at (x, y)
        grad_x, grad_y = grad_f(x, y)
        # Update x and y by taking a step in the opposite direction of the gradient
        x = x - alpha * grad_x
        y = y - alpha * grad_y
    return x, y
```

```python
def fun_1(x, y):
    return x**2 + y**2
```

```python
def grad_f_1(x, y):
    grad_x = 2 * x
    grad_y = 2 * y
    return grad_x, grad_y
```

```python
import numpy as np

def fun_2(x, y):
    return 1 - np.exp(-x**2 - (y - 2)**2) - 2 * np.exp(-x**2 - (y + 2)**2)
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

result_2_case1 = gradient_descent(0, 1, fun_2, grad_f_2, 0.01, 10000)
result_2_case2 = gradient_descent(0, -1, fun_2, grad_f_2, 0.01, 10000)

# Plotting the functions
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(X, Y)
z_1 = fun_1(x, y)
z_2 = fun_2(x, y)

fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(x, y, z_1, cmap='viridis', edgecolor='none')
ax1.set_title('Function f1(x, y) = x^2 + y^2')

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(x, y, z_2, cmap='viridis', edgecolor='none')
ax2.set_title('Function f2(x, y)')

plt.show()
```
