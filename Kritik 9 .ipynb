```python
import sympy as sp

# Defining symbols
x, y = sp.symbols('x y')

# Defining the function given
f = sp.exp(x) * sp.sin(y) + y**3

#1st partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("1st partial derivative w.r.t. x:")
print(df_dx)
print("1st partial derivative w.r.t. y:")
print(df_dy)
```

```python
import sympy as sp


x, y = sp.symbols('x y')

g = x**2 * y + x * y**2

# Gradient vector
gradient_vector = [sp.diff(g, variable) for variable in (x, y)]
print("Gradient vector =>")
print(gradient_vector)

# gradient vector at the point (1, -1)
gradient_magnitude = sp.sqrt(sum(variable**2 for variable in gradient_vector)).subs({x: 1, y: -1})
print("Magnitude of the gradient vector at (1, -1):")
print(gradient_magnitude)
```

```python
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

# Define the function h(x, y) = ln(x^2 + y^2)
h = sp.ln(x**2 + y**2)

# Compute the second partial derivatives
d2h_dx2 = sp.diff(sp.diff(h, x), x)
d2h_dy2 = sp.diff(sp.diff(h, y), y)
d2h_dxdy = sp.diff(sp.diff(h, x), y)

print("2nd partial derivative with respect to xx:")
print(d2h_dx2)
print("2nd partial derivative with respect to yy:")
print(d2h_dy2)
print("2nd partial derivative w.r.t. xy and yx:")
print(d2h_dxdy)
 
print("The symmetry of mixed partials can be explained by Clariaut's Theorem that states: Suppose that f(x,y) is a function. If for a point (a,b) and some open ball around (a,b) then the 2nd order partials exist and are continuous, then fxy(a,b) = fyx(a,b).")
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')


j = x**3 - 3*x*y + y**3

# Converting the SymPy expression to a numerical function
j_func = sp.lambdify((x, y), j, 'numpy')

# Plot grid
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

# Plot function
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
